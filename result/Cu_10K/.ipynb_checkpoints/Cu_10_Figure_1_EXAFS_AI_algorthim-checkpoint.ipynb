```python
from pathlib import Path
import sys,importlib
import os
sys.path.insert(0, '../')


import larch_score
best_Fit  = np.array([[1,0,0.005,-0.05]])
# best_Fit = np.mean(full_mat,axis=0).reshape(-1,4).round(5)

print("Average:")
print(best_Fit)
params = {}
params['Kmin'] = 3.0
params['Kmax'] = 17.0
params['kweight'] = 2
params['deltak'] = 0.05
params['rbkg'] = 1.2
params['bkgkw'] = 1.0
params['bkgkmax'] = 25
params['front'] = "path_files/Cu/path_75/feff"

exp,g,params = larch_score.larch_init(CSV_PATH,params)

path,yTotal,best,loss = larch_score.fitness(exp,best_Fit,paths,params);

SMALL = params['SMALL']
BIG = params['BIG']


# plt.plot(g.k, g.chi*g.k**2,'b-',label="Data",)
plt.plot(path.k[SMALL:BIG], yTotal[SMALL:BIG]*path.k[SMALL:BIG]**2,'r--',label="GA")
plt.legend()


plt.figure()

# plt.plot(g.r,g.chir_mag,'b-',label='exp')
# plt.plot(best.r,best.chir_mag,'r--',label='GA')
# plt.legend()

```

```python
delr = [-0.05,-0.03,-0.01,0.01,0.03,0.05]
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1, 1, 1)

colors =mpl.cm.seismic(np.linspace(0,1,6))

linestyle = ['-','--','-.','']
header_base = 'Cu_10'
name = 'Cu_10_Figure1_Data.csv'

with open(name, mode='w', newline='', encoding='utf-8') as write_file:
    writer = csv.writer(write_file, delimiter=',')

    for i in range(6):
        best_Fit  = np.array([[1,0,0.005,delr[i]]])
        path,yTotal,best,loss = larch_score.fitness(exp,best_Fit,paths,params);
        SMALL = params['SMALL']
        BIG = params['BIG']
        if i == 0 or i == 5:
            ax.plot(path.k[SMALL:BIG], yTotal[SMALL:BIG]*path.k[SMALL:BIG]**2+(1.5*i),color=colors[i],
                label=f"$\Delta r={delr[i]}$")
        elif i == 1 or i == 4:
            ax.plot(path.k[SMALL:BIG], yTotal[SMALL:BIG]*path.k[SMALL:BIG]**2+(1.5*i),'-.',color=colors[i],
            label=f"$\Delta r={delr[i]}$")
        else:
            ax.plot(path.k[SMALL:BIG], yTotal[SMALL:BIG]*path.k[SMALL:BIG]**2+(1.5*i),'--',color=colors[i],
            label=f"$\Delta r={delr[i]}$")

        path_header = ['delr_'+ str(delr[i])+ '_' + header_base +'.k','delr_'+str(delr[i])+'_' + header_base+'.chi2']
        writer.writerow(path_header)
        dat = np.vstack((path.k[SMALL:BIG],yTotal[SMALL:BIG]*path.k[SMALL:BIG]**2+(1.5*i))).T
#         print(dat)
        larch_score.write_dat_csv(writer,dat)
        writer.writerow('')
#             w
# plt.legend()

ax.tick_params(bottom=True, top=True, left=False, right=False)
ax.set_xlim([1,18])
ax.set_ylim([-5,12])
ax.tick_params(which='major',direction='in',width=2,length=10)
ax.tick_params(which='minor',direction='in',width=2,length=7,bottom=True,top=True)
ax.xaxis.set_major_locator(MultipleLocator(2))
ax.xaxis.set_minor_locator(MultipleLocator(0.5))


ax.tick_params(axis='x', which='major', pad=10)
ax.set_yticklabels([]);
ax.legend(bbox_to_anchor=(1.04,0.5), loc="center left")
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[::-1],labels[::-1],bbox_to_anchor=(1.04,0.5), loc="center left")
ax.set_xlabel(r'$k(\AA^{-1})$',fontsize=25)
ax.set_ylabel('Intensity (a.u.)',fontsize=25)
plt.savefig('Cu_10_Fig1.eps',format='eps')
plt.savefig('Cu_10_Fig1.png',dpi=500)
```
