```python
import os
sys.path.insert(0, '../')


import larch_score

importlib.reload(larch_score)

paths = [1,2,5,10,16,28,29,30,31,36,37,40,41,42]


label = larch_score.generate_labels(paths)[0]
num_path = len(paths)
full_mat = []
for i in range(100):
    file = 'test_' + str(i) + '_data.csv'
#     file = 'cu_semi_test_data.csv'
    try:
        os.path.exists(file)
        
        gen_csv = np.genfromtxt(file,delimiter=',')
        gen_csv_unflatten = gen_csv.reshape((-1,4*num_path))
        if i == 0:
            full_mat = gen_csv_unflatten
        full_mat = np.vstack((full_mat,gen_csv_unflatten))
#         print(str(i) + ' Pass')
#         print(full_mat.shape)
    except OSError:
        print(str(i) + " Fail")
        pass
    
```

```python
full_mat_var = np.delete(full_mat, np.s_[5::4],axis=1)
print(full_mat_var.shape)


full_mat_var_cov = np.cov(full_mat_var.T)
print(full_mat_var_cov.shape)
importlib.reload(larch_score)
full_mat_diag = np.diag(full_mat_var_cov)

err = np.sqrt(full_mat_diag)

plt.figure(figsize=(8,5))
plt.xticks(np.arange(len(full_mat_diag)),label,rotation=70);
plt.bar(np.arange(len(full_mat_diag)),err)
plt.title('Cu_10 Cov constructed from ' + str(full_mat_var.shape[0]) + ' observations')
plt.ylabel('Standard Error')
```

```python
from pathlib import Path
import sys,importlib
import os
sys.path.insert(0, '../')


import larch_score

importlib.reload(larch_score)

best_Fit = np.mean(full_mat,axis=0).reshape(-1,4).round(5)

print("Average:")
print(best_Fit)
params = {}
params['Kmin'] = 3.0
params['Kmax'] = 17.0
params['kweight'] = 2
params['deltak'] = 0.05
params['rbkg'] = 1.2
params['bkgkw'] = 1.0
params['bkgkmax'] = 25
params['front'] = "path_files/Cu/path_75/feff"

exp,g,params = larch_score.larch_init(CSV_PATH,params)

path,yTotal,best,loss,best_Fit_r = larch_score.fitness(exp,best_Fit,paths,params,return_r=True);

SMALL = params['SMALL']
BIG = params['BIG']


plt.plot(g.k, g.chi*g.k**2,'b-',label="Data",)
plt.plot(path.k[SMALL:BIG], yTotal[SMALL:BIG]*path.k[SMALL:BIG]**2,'r--',label="GA")
plt.legend()


plt.figure()

plt.plot(g.r,g.chir_mag,'b-',label='exp')
plt.plot(best.r,best.chir_mag,'r--',label='GA')
plt.legend()

```
