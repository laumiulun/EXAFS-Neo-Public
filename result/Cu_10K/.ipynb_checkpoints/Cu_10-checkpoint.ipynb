```python
%matplotlib inline
import numpy as np 
import configparser
import matplotlib.pyplot as plt
config = configparser.ConfigParser()
import os
```

```python
from pathlib import Path
import sys,importlib
sys.path.insert(0, '../')


import larch_score

importlib.reload(larch_score)

base =  Path(os.getcwd()).parent.parent
front = os.path.join(base,"path_files/Cu/feff")

end = '.dat'
CSV_PATH = os.path.join(base,"path_files/Cu/cu_10k.xmu")

paths = np.arange(1,8)
```

```python
label = []
s02_label = []
sigma2_label = []
deltaR_label = []
path_list = np.arange(8)
for i in range(8):
    label.append('s02_' + str(path_list[i]+1))
    if i == 0:
        label.append('e0')

    s02_label.append('s02_' + str(path_list[i]+1))
#     label.append('e0_' + str(path_list[i]))
    label.append('sigma_' + str(path_list[i]+1))
    sigma2_label.append('sigma_' + str(path_list[i]+1))
    
    label.append('deltaR_' + str(path_list[i]+1))
    deltaR_label.append('deltaR_' + str(path_list[i]+1))
```

```python
num_path = 7
full_mat = []
for i in range(100):
    file = 'test_' + str(i) + '_data.csv'
    
    try:
        os.path.exists(file)
        
        gen_csv = np.genfromtxt(file,delimiter=',')
        gen_csv_unflatten = gen_csv.reshape((-1,num_path*4))
        if i == 0:
            full_mat = gen_csv_unflatten
        full_mat = np.vstack((full_mat,gen_csv_unflatten))
        print(str(i) + ' Pass')
        print(full_mat.shape)
    except OSError:
        print(str(i) + " Fail")
        pass
    
```

```python
full_mat_var = np.delete(full_mat, np.s_[5::4],axis=1)
print(full_mat_var.shape)


full_mat_var_cov = np.cov(full_mat_var.T)
print(full_mat_var_cov.shape)

full_mat_diag = np.diag(full_mat_var_cov)


plt.xticks(np.arange(len(full_mat_diag)),label,rotation=70);
plt.bar(np.arange(len(full_mat_diag)),np.sqrt(full_mat_diag))
plt.title('Cu_10K Cov constructed from ' + str(full_mat_var.shape[0]) + ' observations')
plt.ylabel('Standard Error')
```

```python
from pathlib import Path
import sys,importlib
import os
sys.path.insert(0, '../')


import larch_score

importlib.reload(larch_score)

best_Fit = np.mean(full_mat,axis=0).reshape(-1,4).round(5)
print("Average:")
print(best_Fit)
params = {}
params['Kmin'] = 3
params['Kmax'] = 17
params['kweight'] = 2
params['deltak'] = 0.05
params['rbkg'] = 1.2
params['bkgkw'] = 1
params['bkgkmax'] = 25
params['front'] = "path_files/Cu/path_7_jeff/feff"

exp,g,params = larch_score.larch_init(CSV_PATH,params)

path,yTotal,best,loss = larch_score.fitness(exp,best_Fit,paths,params);

SMALL = params['SMALL']
BIG = params['BIG']


plt.plot(g.k, g.chi*g.k**2,label="Data")
plt.plot(path.k[SMALL:BIG], yTotal[SMALL:BIG]*path.k[SMALL:BIG]**2,label="GA")
plt.legend()


plt.figure()

plt.plot(g.r,g.chir_mag,label='exp')
plt.plot(best.r,best.chir_mag,label='GA')
plt.legend()
```
